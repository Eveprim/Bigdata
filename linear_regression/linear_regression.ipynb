# 2. 가설함수 정의

```python
W = np.ones((3,1))
```

```python
b = np.ones((1,1))
```

```python
def f(X):
    return np.matmul(X, W) + b
```

```python
y_pred = f(x_train)
```

# 3. cost 함수 정의

```python
def loss_function(y_pred, y_train):
    n_samples = y_train.shape[0]
    deviation = y_pred - y_train
    deviation_squared = deviation**2
    loss = (1 / n_samples) * np.sum(deviation_squared)

    return loss
```

```python
loss_function(y_pred, y_train)
```

```python
def gradient(x_train, y_train, y_pred):
    n_samples = y_train.shape[0]
    dw = (1 / n_samples) * np.matmul(x_train.transpose(), (y_pred - y_train))
    db = (1/n_samples) * np.sum(y_pred - y_train, axis=0, keepdims=True)

    return dw, db
```

```python
def gradient_descent(W, b, learning_rate, dw, db):
    W = W - learning_rate * dw
    b = b - learning_rate * db
    
    return W, b
```

```python
dw, db = gradient(x_train, y_train, y_pred)
```

```python
gradient_descent(W, b, 0.01, dw, db)
```
